# Pipeline Xử lý Dữ liệu v2.0: Xây dựng Master Dataset Tin cậy

**Mục tiêu:** Notebook này trình bày một quy trình tiền xử lý dữ liệu minh bạch và được ghi chú đầy đủ, nhằm giải quyết các thắc mắc và xây dựng một nguồn dữ liệu (`master_data.csv`) đáng tin cậy cho toàn bộ team Data.

**Quy trình:**
1.  **Tải Dữ liệu & Khám phá ban đầu:** Đánh giá chất lượng dữ liệu thô.
2.  **Làm sạch và Chuẩn hóa:** Xử lý từng file với các bước kiểm tra tính toàn vẹn và ghi chú chi tiết.
3.  **Biện minh cho các Quyết định Xử lý:** Giải thích lý do lựa chọn các phương pháp xử lý giá trị thiếu và chiến lược hợp nhất.
4.  **Hợp nhất và Xuất file:** Tạo ra file `master_data_v2.csv` cuối cùng.

## 1. Tải Dữ liệu và Khám phá ban đầu

```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns', None)

# Tải dữ liệu thô
prod_raw = pd.read_csv('../data/raw/production_logs.csv', low_memory=False)
maint_raw = pd.read_csv('../data/raw/maintenance_order.csv')
cross_ref_raw = pd.read_csv('../data/raw/cross_reference.csv')
```

## 2. Tiền xử lý (Preprocessing) & Biện minh

### 2.1. Xử lý `cross_reference.csv` (Cầu nối Dữ liệu)
**Mục tiêu:** Đảm bảo đây là một bảng tra cứu (lookup table) 1-1 sạch sẽ.

```python
print("--- Xử lý cross_reference.csv ---")
cross_ref = cross_ref_raw.copy()

# Đổi tên cột cho nhất quán
cross_ref.rename(columns={'OPERA NAME': 'LINE_NAME'}, inplace=True)

# **KIỂM TRA TÍNH TOÀN VẸN 1: Loại bỏ các cặp (ID, NAME) trùng lặp**
print(f"Số dòng trước khi drop_duplicates: {len(cross_ref)}")
cross_ref.drop_duplicates(inplace=True)
print(f"Số dòng sau khi drop_duplicates: {len(cross_ref)}")

# **KIỂM TRA TÍNH TOÀN VẸN 2: Đảm bảo mỗi EQUIPMENT_ID là duy nhất**
# Đây là bước kiểm tra quan trọng nhất để tránh lỗi khi merge.
if cross_ref['EQUIPMENT_ID'].duplicated().any():
    print("CẢNH BÁO NGHIÊM TRỌNG: Tồn tại EQUIPMENT_ID bị trùng lặp. Điều này có nghĩa là một máy đang được map tới nhiều tên dây chuyền khác nhau.")
    duplicated_ids = cross_ref[cross_ref['EQUIPMENT_ID'].duplicated()]['EQUIPMENT_ID']
    print(f"Các ID bị trùng: {duplicated_ids.tolist()}")
else:
    print("KIỂM TRA THÀNH CÔNG: Mỗi EQUIPMENT_ID là duy nhất. Bảng tra cứu hợp lệ.")
```

### 2.2. Xử lý `maintenance_order.csv`

```python
print("\n--- Xử lý maintenance_order.csv ---")
maint = maint_raw.copy()

# Chuyển đổi kiểu dữ liệu cho cột ngày
maint['BASIC_START_DATE'] = pd.to_datetime(maint['BASIC_START_DATE'])

# Loại bỏ các dòng trùng lặp hoàn toàn (cùng một lệnh bảo trì được ghi nhận nhiều lần)
print(f"Số dòng trước khi drop_duplicates: {len(maint)}")
maint.drop_duplicates(inplace=True)
print(f"Số dòng sau khi drop_duplicates: {len(maint)}")
```

### 2.3. Xử lý `production_logs.csv` (File chính)
Đây là phần quan trọng nhất, nơi các quyết định xử lý cần được biện minh rõ ràng.

```python
prod_raw.shape
```

```python
# Đếm tổng số dòng trùng hoàn toàn
duplicate_count = prod_raw.duplicated(keep=False).sum()
print(f"Số dòng trùng hoàn toàn: {duplicate_count}")

# Nếu muốn xem chi tiết các dòng trùng
duplicate_rows = prod_raw[prod_raw.duplicated(keep=False)]
print("\nCác dòng trùng hoàn toàn:")
print(duplicate_rows)
```

```python
# 2.3. Xử lý production_logs.csv (File chính)
# Ở phiên bản này, chúng ta sẽ thực hiện các bước xử lý tối thiểu để chuẩn bị cho việc merge.
# Việc xử lý null và trùng lặp sẽ được thực hiện SAU KHI MERGE.

print("\n--- Xử lý production_logs.csv (Chuẩn bị để Merge) ---")
prod = prod_raw.copy()

# 1. Chuẩn hóa Cột và Kiểu dữ liệu
prod['PRODUCTIONDATE_DAY_LOC'] = pd.to_datetime(prod['PRODUCTIONDATE_DAY_LOC'])
if 'SHIFT_NAME.1' in prod.columns:
    prod.drop(columns=['SHIFT_NAME.1'], inplace=True)

print(f"Kích thước của production_logs sau khi chuẩn bị: {prod.shape}")
print("Lưu ý: Chưa loại bỏ trùng lặp hoặc xử lý null ở bước này.")
```

## 3. Hợp nhất thành Master Dataset
**Biện minh cho Chiến lược Hợp nhất:**

- **Loại Merge:** Chúng ta sẽ sử dụng `how='left'` với `prod` là bảng bên trái.
- **Lý do:** Bảng `production_logs` là **nguồn chân lý (source of truth)** cho mọi hoạt động của nhà máy. Chúng ta phải giữ lại tất cả các bản ghi trong file này. Thông tin từ file bảo trì (`maintenance`) được xem như là thông tin bổ sung (enrichment), được thêm vào những ngày có sự kiện bảo trì tương ứng. Nếu một ngày không có bảo trì, các cột liên quan sẽ là `NaN`, và chúng ta sẽ xử lý nó bằng cách điền `'No Maintenance'`.

```python
# Bước 1: Gắn 'LINE_NAME' vào dữ liệu bảo trì
maint_enriched = pd.merge(maint, cross_ref, on='EQUIPMENT_ID', how='left')
maint_enriched.rename(columns={'BASIC_START_DATE': 'PRODUCTIONDATE_DAY_LOC'}, inplace=True)
```

```python
maint_enriched.shape
```

```python
# Bước 2: Hợp nhất dữ liệu sản xuất và bảo trì
master_df = pd.merge(prod,
                     maint_enriched[['PRODUCTIONDATE_DAY_LOC', 'LINE_NAME', 'EQUIPMENT_ID', 'ORDER_TYPE', 'DESCRIPTION', 'ORDER']],
                     on=['PRODUCTIONDATE_DAY_LOC', 'LINE_NAME'],
                     how='left')

# Bước 3: Xử lý các giá trị NaN phát sinh sau khi merge
master_df['ORDER_TYPE'].fillna('No Maintenance Event', inplace=True)
master_df['DESCRIPTION'].fillna('No Maintenance Event', inplace=True)
master_df['ORDER'].fillna('No Maintenance Event', inplace=True)

# Do có thể có nhiều lệnh bảo trì trên cùng 1 line trong 1 ngày, việc merge có thể tạo ra các dòng trùng lặp
# Cần phải loại bỏ chúng
master_df.drop_duplicates(inplace=True)

print(f"Kích thước của Master DataFrame sau khi hợp nhất và xử lý trùng lặp: {master_df.shape}")
```

## 4. Kiểm tra cuối cùng và Xuất file

```python
print("--- Kiểm tra 5 dòng đầu của Master Dataset ---")
display(master_df.head())

print("--- Thông tin tổng quan của Master Dataset ---")
master_df.info()

# Xuất ra file CSV
output_filename = '../data/processed/master_data.csv'
master_df.to_csv(output_filename, index=False, encoding='utf-8')
print(f"\nPipeline hoàn tất! Dữ liệu đã được lưu vào file '{output_filename}'.")
```
