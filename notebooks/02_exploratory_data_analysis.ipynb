## Phần A: Chuẩn bị Dữ liệu (Data Preparation) - Nền tảng của Phân tích Chính xác

### A1. Tải và Hợp nhất Dữ liệu

Mọi phân tích đều bắt đầu từ một bộ dữ liệu sạch sẽ và đầy đủ. Nhiệm vụ đầu tiên của chúng ta là kết hợp 3 file CSV riêng lẻ thành một DataFrame duy nhất, thống nhất.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập môi trường làm việc
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'figure.figsize': (16, 9),
    'font.size': 14,
    'axes.titlesize': 20,
    'axes.labelsize': 16
})
pd.set_option('display.max_columns', None)

print("Bắt đầu quá trình tải và chuẩn bị dữ liệu...")

# Tải 3 file dữ liệu
production_df = pd.read_csv('../data/raw/production_logs.csv', low_memory=False)
maintenance_df = pd.read_csv('../data/raw/maintenance_order.csv')
cross_ref_df = pd.read_csv('../data/raw/cross_reference.csv')

print("--- Kích thước ban đầu ---")
print(f"Production Logs: {production_df.shape}")
print(f"Maintenance Orders: {maintenance_df.shape}")
print(f"Cross Reference: {cross_ref_df.shape}")

# Bước 1: Chuẩn bị bảng tham chiếu (cross_reference)
# Mục đích: Liên kết mã thiết bị vật lý (EQUIPMENT_ID) với tên dây chuyền vận hành (LINE_NAME).
cross_ref_df.rename(columns={'OPERA NAME': 'LINE_NAME'}, inplace=True)
# Lập luận: Một EQUIPMENT_ID chỉ nên tương ứng với một LINE_NAME. Việc loại bỏ trùng lặp đảm bảo tính toàn vẹn khi hợp nhất.
cross_ref_df.drop_duplicates(subset=['EQUIPMENT_ID'], inplace=True)

# Bước 2: Làm giàu dữ liệu bảo trì (maintenance_order)
# Mục đích: Thêm tên dây chuyền (LINE_NAME) vào dữ liệu bảo trì để có thể hợp nhất với dữ liệu sản xuất.
maintenance_full_df = pd.merge(maintenance_df, cross_ref_df, on='EQUIPMENT_ID', how='left')

# Bước 3: Hợp nhất dữ liệu sản xuất và bảo trì
# Lập luận: Chúng ta cần một dòng thời gian duy nhất. Việc hợp nhất (merge) dựa trên ngày và tên dây chuyền cho phép chúng ta xem xét các sự kiện sản xuất và bảo trì xảy ra trên cùng một máy, trong cùng một ngày.
# Chuyển đổi cột ngày tháng sang định dạng datetime để hợp nhất chính xác.
production_df['PRODUCTIONDATE_DAY_LOC'] = pd.to_datetime(production_df['PRODUCTIONDATE_DAY_LOC'])
maintenance_full_df['BASIC_START_DATE'] = pd.to_datetime(maintenance_full_df['BASIC_START_DATE'])
maintenance_full_df.rename(columns={'BASIC_START_DATE': 'PRODUCTIONDATE_DAY_LOC'}, inplace=True)

# Sử dụng 'left' merge để giữ lại tất cả các log sản xuất, là dữ liệu chính của chúng ta.
df_master = pd.merge(production_df, maintenance_full_df, on=['PRODUCTIONDATE_DAY_LOC', 'LINE_NAME'], how='left')

print(f"\nKích thước sau khi hợp nhất: {df_master.shape}")
print("Kiểm tra 5 dòng đầu của bộ dữ liệu đã hợp nhất:")
display(df_master.head())
```

### A2. Làm sạch và Tạo các Biến mới (Feature Engineering)

Dữ liệu thô hiếm khi hoàn hảo. Bước này cực kỳ quan trọng để đảm bảo tính chính xác và tạo ra các biến mới hữu ích cho phân tích.

```python
print("--- Quá trình làm sạch và Feature Engineering ---")

# 1. Xử lý trùng lặp
# Lập luận: Các dòng trùng lặp hoàn toàn có thể làm sai lệch các phép tính tổng và đếm. Cần loại bỏ chúng.
initial_rows = len(df_master)
df_master.drop_duplicates(inplace=True)
print(f"Đã loại bỏ {initial_rows - len(df_master)} dòng trùng lặp.")

# 2. Xử lý các cột không cần thiết hoặc bị lỗi
# Lập luận: Cột 'SHIFT_NAME.1' có vẻ là một lỗi khi xuất dữ liệu, chứa thông tin trùng lặp. Loại bỏ nó giúp DataFrame gọn gàng hơn.
if 'SHIFT_NAME.1' in df_master.columns:
    df_master.drop('SHIFT_NAME.1', axis=1, inplace=True)
    print("Đã loại bỏ cột 'SHIFT_NAME.1' bị thừa.")

# 3. Xử lý giá trị thiếu (Missing Values)
# Lập luận: PRODUCTION_ORDER_RATE là biến số quan trọng để tính toán hiệu suất. Nếu để trống (NaN), các phép tính sẽ bị lỗi.
# Chúng ta điền 0 với giả định rằng nếu không có tốc độ đặt hàng, máy đang không trong trạng thái sản xuất có thể đo lường được.
df_master['PRODUCTION_ORDER_RATE'].fillna(0, inplace=True)
print("Đã xử lý giá trị thiếu cho 'PRODUCTION_ORDER_RATE'.")

# 4. Tạo các biến mới (Feature Engineering)
# Lập luận: Việc tạo ra các biến mới giúp đơn giản hóa và làm rõ các phân tích sau này.
# - TOTAL_PRODUCTION_QTY: Tổng sản lượng, cần thiết để tính tỷ lệ lỗi.
# - MONTH, WEEKDAY: Các biến thời gian giúp phân tích xu hướng và các mẫu theo chu kỳ (ví dụ: thứ Hai có tệ hơn thứ Sáu không?).
df_master['TOTAL_PRODUCTION_QTY'] = df_master['GOOD_PRODUCTION_QTY'] + df_master['REJECT_PRODUCTION_QTY']
df_master['MONTH'] = df_master['PRODUCTIONDATE_DAY_LOC'].dt.month
df_master['WEEKDAY'] = df_master['PRODUCTIONDATE_DAY_LOC'].dt.day_name()
df_master['IS_PM_DAY'] = ~df_master['ORDER_TYPE'].str.contains('PM', na=False).isnull()
print("Đã tạo các biến mới: 'TOTAL_PRODUCTION_QTY', 'MONTH', 'WEEKDAY', 'IS_PM_DAY'.")

print("\n--- Tổng quan về dữ liệu sau khi chuẩn bị ---")
print(f"Kích thước cuối cùng: {df_master.shape}")
print("Thông tin các cột và kiểu dữ liệu:")
df_master.info()
```

## Phần B: Phân tích Khám phá (EDA) - Kể chuyện bằng Dữ liệu

### B1. Bức tranh Toàn cảnh: Nhà máy đang hoạt động như thế nào?
Trước khi đi vào chi tiết, chúng ta cần một cái nhìn tổng thể. Thời gian của nhà máy được sử dụng vào việc gì? Và tổn thất tài chính lớn đến mức nào?

```python
print("--- Bảng 1.1: Phân Bổ Tổng Thời Gian Hoạt Động ---")
time_cols = ['RUN_TIME', 'DOWNTIME', 'PLANNED_MAINTENANCE', 'CHANGEOVER_DURATION', 'CLEANING_AND_SANITIZATION', 'NO_DEMAND', 'LUNCH_AND_BREAK']
time_totals = df_master[time_cols].sum().reset_index()
time_totals.columns = ['Hạng mục Thời gian', 'Tổng Thời gian (phút)']
time_totals['Tỷ lệ (%)'] = (time_totals['Tổng Thời gian (phút)'] / time_totals['Tổng Thời gian (phút)'].sum()) * 100
time_totals = time_totals.sort_values(by='Tổng Thời gian (phút)', ascending=False)
display(time_totals)

plt.figure(figsize=(12, 12))
plt.pie(time_totals['Tỷ lệ (%)'], labels=time_totals['Hạng mục Thời gian'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4), pctdistance=0.8)
plt.title('Phân Bổ Thời Gian Hoạt Động: Gần 40% Thời gian bị lãng phí', fontsize=22, fontweight='bold', pad=20)
plt.show()

print("\n--- Bảng 1.3: Định Lượng Sơ bộ Tác Động Kinh Doanh ---")
df_master['LOST_UNITS_DOWNTIME'] = df_master['DOWNTIME'] * df_master['PRODUCTION_ORDER_RATE']
impact_summary = {
    'Hạng mục': ['Sản lượng mất do Downtime (đơn vị)', 'Sản phẩm lỗi (đơn vị)', 'TỔN THẤT TÀI CHÍNH ƯỚC TÍNH ($)'],
    'Giá trị': [
        f"{int(df_master['LOST_UNITS_DOWNTIME'].sum()):,}",
        f"{int(df_master['REJECT_PRODUCTION_QTY'].sum()):,}",
        f"{(df_master['LOST_UNITS_DOWNTIME'].sum() * 0.5 + df_master['REJECT_PRODUCTION_QTY'].sum() * 0.15):,.2f}"
    ]
}
display(pd.DataFrame(impact_summary))
```

**Diễn giải & Insight ban đầu:**
- **Thời gian là Vàng, và chúng ta đang lãng phí nó:** Biểu đồ tròn cho thấy một sự thật đáng báo động. Chỉ **61.7%** tổng thời gian là `RUN_TIME` (thời gian máy thực sự chạy để tạo ra sản phẩm). Gần **40%** thời gian còn lại là lãng phí.
- **Hai "Kẻ cắp" Thời gian lớn nhất:** `NO_DEMAND` (18.8%) và `DOWNTIME` (11.3%) là hai nguyên nhân lớn nhất. `NO_DEMAND` là một vấn đề về hoạch định và kinh doanh, cho thấy năng lực nhà máy chưa được khai thác hết. `DOWNTIME` là vấn đề vận hành cốt lõi mà chúng ta cần mổ xẻ.
- **Con số biết nói:** Tổn thất **41.3 triệu đô la** là một lý do đủ lớn để mọi người phải ngồi lại và nghiêm túc xem xét các vấn đề này. Nó cho thấy việc cải thiện hiệu suất không chỉ là vấn đề kỹ thuật, mà là một yêu cầu sống còn về mặt kinh doanh.

### B2. Mổ xẻ Downtime: Ai, Cái gì, Ở đâu, Khi nào?
Bây giờ, chúng ta sẽ đi sâu vào khoản lãng phí 11.3% do `DOWNTIME`. Chúng ta cần trả lời các câu hỏi: Nguyên nhân nào là lớn nhất? Dây chuyền nào là "điểm nóng"?

```python
print("--- Bảng 2.1: Top 15 Nguyên Nhân Gây Downtime (Theo Tổng thời gian & Tần suất) ---")
reason_summary = df_master[df_master['DOWNTIME'] > 0].groupby('UTIL_REASON_DESCRIPTION')['DOWNTIME'].agg(['sum', 'count']).reset_index()
reason_summary.columns = ['Nguyên nhân', 'Tổng Downtime (phút)', 'Số lần xảy ra']
reason_summary = reason_summary.sort_values(by='Tổng Downtime (phút)', ascending=False).head(15)
display(reason_summary)

fig, ax = plt.subplots(2, 1, figsize=(18, 16))
sns.barplot(x='Tổng Downtime (phút)', y='Nguyên nhân', data=reason_summary.head(10), palette='viridis', ax=ax[0])
ax[0].set_title('Top 10 Nguyên Nhân Gây Downtime (Theo Tổng thời gian)', fontweight='bold')
ax[0].set_xlabel('Tổng Thời Gian Downtime (phút)')

sns.barplot(x='Số lần xảy ra', y='Nguyên nhân', data=reason_summary.sort_values(by='Số lần xảy ra', ascending=False).head(10), palette='plasma', ax=ax[1])
ax[1].set_title('Top 10 Nguyên Nhân Gây Downtime (Theo Tần suất)', fontweight='bold')
ax[1].set_xlabel('Số lần xảy ra')

plt.tight_layout()
plt.show()
```

**Diễn giải & Insight:**
- **Vấn đề kép:** Chúng ta có hai loại vấn đề chính:
    1.  **Sự cố Lớn (High Duration):** `02-Quality - 0202-Rework` đứng đầu tuyệt đối về tổng thời gian downtime. Mặc dù nó chỉ xảy ra ~3200 lần, nhưng mỗi lần xảy ra đều rất tốn thời gian. Đây là những "vết thương" lớn, cần được ưu tiên chữa trị.
    2.  **Sự cố Hệ thống (High Frequency):** `17-End Feeder - 1702-Jammed handle on chain` là sự cố xảy ra thường xuyên nhất (>4000 lần). Mỗi lần có thể chỉ mất vài phút, nhưng sự lặp lại liên tục của nó gây gián đoạn và bào mòn hiệu suất hàng ngày. Đây là những "cái gai" nhỏ nhưng khó chịu, cần các giải pháp cải tiến liên tục (Kaizen).

```python
print("--- Bảng 2.2: Top 10 Dây Chuyền Có Downtime Cao Nhất - Mặt trận Sẵn sàng ---")
line_summary = df_master.groupby('LINE_NAME')['DOWNTIME'].sum().reset_index()
line_summary = line_summary.sort_values(by='DOWNTIME', ascending=False).head(10)
display(line_summary)

sns.barplot(x='DOWNTIME', y='LINE_NAME', data=line_summary, palette='magma')
plt.title('Top Dây Chuyền có Downtime cao: Sự thống trị của nhóm MKZA', fontweight='bold')
plt.xlabel('Tổng Thời Gian Downtime (phút)')
plt.ylabel('Dây chuyền')
plt.show()
```

**Diễn giải & Insight:**
- **"Điểm nóng" MKZA:** Dữ liệu chỉ ra một cách rõ ràng rằng nhóm dây chuyền `MKZA` (1501, 1502, 1601, 1402, 1802, 1401, 1801) là nguồn gốc chính của downtime. Có tới 7/10 dây chuyền trong top đầu thuộc nhóm này. Điều này cho thấy có thể có vấn đề chung về thiết kế, tuổi thọ, hoặc quy trình vận hành/bảo trì cho cả nhóm dây chuyền này.

```python
print("--- Bảng 2.3: Phân Tích Chéo - Sự cố Downtime xảy ra ở đâu? ---")
top_lines = line_summary['LINE_NAME'].tolist()
top_reasons = reason_summary.head(5)['Nguyên nhân'].tolist()
crosstab_downtime = pd.crosstab(index=df_master[df_master['LINE_NAME'].isin(top_lines)]['LINE_NAME'],
                                columns=df_master[df_master['UTIL_REASON_DESCRIPTION'].isin(top_reasons)]['UTIL_REASON_DESCRIPTION'],
                                values=df_master['DOWNTIME'], aggfunc='sum').fillna(0).astype(int)
display(crosstab_downtime)

plt.figure(figsize=(18, 10))
sns.heatmap(crosstab_downtime, annot=True, fmt='d', cmap='viridis', linewidths=.5)
plt.title('Heatmap Downtime: Vấn đề Rework tập trung ở đâu?', fontweight='bold')
plt.show()
```

**Diễn giải & Insight:**
- **Câu chuyện được làm rõ:** Heatmap này kết nối các điểm lại với nhau. Nó cho thấy vấn đề `Rework` (cột đầu tiên) xảy ra nhiều nhất trên chính các dây chuyền `MKZA` (1601, 1801, 1402, 1502, 1501).
- **Giả thuyết mới:** Có thể các sự cố cơ khí trên dây chuyền `MKZA` không chỉ gây downtime, mà còn **dẫn đến sản phẩm lỗi**, và sau đó lại phải downtime thêm một lần nữa để `Rework`. Đây là một **vòng lặp tiêu cực** cần được phá vỡ.

### B3. Mổ xẻ Chất lượng: Ai đang tạo ra phế phẩm?
Bây giờ chúng ta chuyển sang phân tích `REJECT_PRODUCTION_QTY`. Dây chuyền nào, sản phẩm nào là nguồn gốc chính của lãng phí nguyên vật liệu?

```python
print("--- Bảng 3.1 & 3.2: Nguồn gốc của Sản phẩm lỗi ---")
total_rejects = df_master['REJECT_PRODUCTION_QTY'].sum()

reject_line = df_master.groupby('LINE_NAME')['REJECT_PRODUCTION_QTY'].sum().reset_index()
reject_line['Tỷ lệ (%)'] = (reject_line['REJECT_PRODUCTION_QTY'] / total_rejects) * 100
reject_line = reject_line.sort_values(by='REJECT_PRODUCTION_QTY', ascending=False).head(10)

reject_product = df_master.groupby('SIZE_TYPE')['REJECT_PRODUCTION_QTY'].sum().reset_index()
reject_product['Tỷ lệ (%)'] = (reject_product['REJECT_PRODUCTION_QTY'] / total_rejects) * 100
reject_product = reject_product.sort_values(by='REJECT_PRODUCTION_QTY', ascending=False).head(10)

print("Top 10 Dây Chuyền Gây Lỗi Nhiều Nhất:")
display(reject_line)

print("\nTop 10 Loại Sản Phẩm Có Nhiều Lỗi Nhất:")
display(reject_product)

fig, ax = plt.subplots(1, 2, figsize=(24, 10))
sns.barplot(x='REJECT_PRODUCTION_QTY', y='LINE_NAME', data=reject_line, palette='rocket', ax=ax[0])
ax[0].set_title('Top 10 Dây Chuyền Gây Lỗi Nhiều Nhất', fontweight='bold')
sns.barplot(x='REJECT_PRODUCTION_QTY', y='SIZE_TYPE', data=reject_product, palette='mako', ax=ax[1])
ax[1].set_title('Top 10 Sản Phẩm Có Nhiều Lỗi Nhất', fontweight='bold')
plt.show()
```

**Diễn giải & Insight:**
- **Mặt trận Chất lượng:** Không giống như downtime, vấn đề chất lượng lại tập trung ở một nhóm dây chuyền hoàn toàn khác: `MKBC` (0901, 1101, 1002, 0902) và `MHPT` (2303, 2302). Điều này xác nhận giả thuyết **"Cuộc chiến trên hai mặt trận"** của chúng ta.
- **"Sản phẩm Rủi ro cao":** `Classic Clean 100` và `Total Defense Shield` chiếm tới hơn **31%** tổng số sản phẩm lỗi. Quy trình sản xuất của hai sản phẩm này cần được xem xét lại một cách cẩn thận.

```python
print("--- Bảng 3.3: Phân Tích Chéo - Giao điểm chết người giữa Dây chuyền và Sản phẩm ---")
top_lines_reject = reject_line['LINE_NAME'].tolist()
top_products_reject = reject_product['SIZE_TYPE'].tolist()
crosstab_df = df_master[df_master['LINE_NAME'].isin(top_lines_reject) & df_master['SIZE_TYPE'].isin(top_products_reject)]
reject_crosstab = pd.crosstab(index=crosstab_df['LINE_NAME'], columns=crosstab_df['SIZE_TYPE'],
                                  values=crosstab_df['REJECT_PRODUCTION_QTY'], aggfunc='sum').fillna(0).astype(int)
display(reject_crosstab)

plt.figure(figsize=(20, 12))
sns.heatmap(reject_crosstab, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5, annot_kws={"size": 12})
plt.title('Heatmap Lỗi: Nơi các vấn đề hội tụ', fontweight='bold')
plt.show()
```

**Diễn giải & Insight:**
- **Bằng chứng không thể chối cãi:** Heatmap này là công cụ mạnh mẽ nhất để xác định "điểm nóng" chất lượng. Nó chỉ ra chính xác rằng:
    - Dây chuyền `MKBC1101` có vấn đề nghiêm trọng với sản phẩm `Classic Clean 100`.
    - Dây chuyền `MHPT2302` có vấn đề nghiêm trọng với `Deep Action White`.
    - Dây chuyền `MHPT2303` có vấn đề nghiêm trọng với `Total Action Model S`.
- **Hành động cụ thể:** Thay vì điều tra chung chung, chúng ta có thể cử các đội cải tiến đến chính xác các giao điểm này để phân tích và giải quyết.

### B4. Yếu tố Con người và Vận hành
Máy móc và quy trình là một phần của câu chuyện. Phần còn lại là con người. Liệu có sự khác biệt về hiệu suất giữa các đội ngũ và các ca làm việc không?

```python
print("--- Bảng 4.1 & 4.2: Phân tích Hiệu suất theo Đội ngũ và Ca làm việc ---")
crew_analysis = df_master.groupby('CREW_ID').agg(
    total_downtime=('DOWNTIME', 'sum'),
    mean_downtime_per_entry=('DOWNTIME', 'mean'),
    total_rejects=('REJECT_PRODUCTION_QTY', 'sum'),
    mean_rejects_per_entry=('REJECT_PRODUCTION_QTY', 'mean')
).sort_values(by='total_downtime', ascending=False)

shift_analysis = df_master.groupby('SHIFT_NAME').agg(
    total_downtime=('DOWNTIME', 'sum'),
    mean_downtime_per_entry=('DOWNTIME', 'mean'),
    total_rejects=('REJECT_PRODUCTION_QTY', 'sum'),
    mean_rejects_per_entry=('REJECT_PRODUCTION_QTY', 'mean')
).sort_values(by='total_downtime', ascending=False)

print("Phân tích theo Đội ngũ (CREW_ID):")
display(crew_analysis)

print("\nPhân tích theo Ca làm việc (SHIFT_NAME):")
display(shift_analysis)

fig, ax = plt.subplots(2, 2, figsize=(24, 16))
sns.barplot(x=crew_analysis.index, y=crew_analysis['total_downtime'], ax=ax[0,0], palette='crest').set_title('Tổng Downtime theo Đội ngũ')
sns.barplot(x=crew_analysis.index, y=crew_analysis['total_rejects'], ax=ax[0,1], palette='crest').set_title('Tổng Sản phẩm lỗi theo Đội ngũ')
sns.barplot(x=shift_analysis.index, y=shift_analysis['total_downtime'], ax=ax[1,0], palette='flare').set_title('Tổng Downtime theo Ca')
sns.barplot(x=shift_analysis.index, y=shift_analysis['total_rejects'], ax=ax[1,1], palette='flare').set_title('Tổng Sản phẩm lỗi theo Ca')
plt.tight_layout()
plt.show()
```

**Diễn giải & Insight:**
- **Không có sự khác biệt lớn:** Một phát hiện khá bất ngờ. Dữ liệu cho thấy tổng downtime và tổng sản phẩm lỗi **khá tương đồng** giữa các đội (A, B, C) và các ca (Day, Afternoon, Night).
- **Hàm ý:** Điều này cho thấy các vấn đề chúng ta đang đối mặt có khả năng là **vấn đề hệ thống** (liên quan đến máy móc, quy trình) hơn là do năng lực của một đội hay một ca cụ thể. Đây là một tin tốt, vì việc sửa chữa hệ thống thường dễ dàng và bền vững hơn là việc đào tạo lại con người.

### B5. Phân tích Nâng cao: Xu hướng theo Thời gian và Hiệu quả Bảo trì
Cuối cùng, chúng ta xem xét dữ liệu theo chiều thời gian để trả lời hai câu hỏi quan trọng: Vấn đề đang tốt lên hay xấu đi? Và các hoạt động bảo trì hiện tại có hiệu quả không?

```python
print("--- Bảng 5.1: Xu hướng Downtime và Lỗi hàng tháng (Tổng thể) ---")
df_ts = df_master.set_index(pd.to_datetime(df_master['PRODUCTIONDATE_DAY_LOC']))
monthly_trends = df_ts.groupby(pd.Grouper(freq='ME'))[['DOWNTIME', 'REJECT_PRODUCTION_QTY']].sum()
monthly_trends.index = monthly_trends.index.strftime('%Y-%m')
display(monthly_trends)

monthly_trends.plot(kind='line', marker='o', subplots=True, layout=(2,1), figsize=(15, 12), sharex=True)
plt.suptitle('Xu Hướng Downtime và Sản phẩm lỗi Hàng Tháng', fontsize=22, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

**Diễn giải & Insight:**
- **Câu chuyện hai nửa:** Biểu đồ cho thấy một câu chuyện rõ ràng. Trong 3 tháng đầu năm (tháng 1-3), cả downtime và sản phẩm lỗi đều có xu hướng **tăng lên**, đạt đỉnh vào tháng 3. Tuy nhiên, từ tháng 4 đến tháng 6, cả hai chỉ số đều có xu hướng **giảm mạnh**.
- **Hành động đã có hiệu quả:** Sự sụt giảm này không thể là ngẫu nhiên. Nó cho thấy rằng một hoặc nhiều hành động cải tiến đã được thực hiện vào khoảng cuối tháng 3, đầu tháng 4 và đã mang lại kết quả tích cực. Nhiệm vụ của chúng ta là tìm ra đó là những hành động gì để nhân rộng.

```python
print("--- Bảng 5.2: Tóm tắt Hiệu quả của Bảo trì Phòng ngừa (PM) ---")
pm_events = df_master[df_master['ORDER_TYPE'].str.contains('PM', na=False)][['PRODUCTIONDATE_DAY_LOC', 'LINE_NAME']].drop_duplicates()
if not pm_events.empty:
    results = []
    events_to_analyze = pm_events.sample(min(100, len(pm_events)), random_state=42)
    for _, row in events_to_analyze.iterrows():
        pm_date, line = row['PRODUCTIONDATE_DAY_LOC'], row['LINE_NAME']
        downtime_before = df_master[(df_master['LINE_NAME'] == line) & (df_master['PRODUCTIONDATE_DAY_LOC'] >= (pm_date - pd.Timedelta(days=7))) & (df_master['PRODUCTIONDATE_DAY_LOC'] < pm_date)]['DOWNTIME'].sum()
        downtime_after = df_master[(df_master['LINE_NAME'] == line) & (df_master['PRODUCTIONDATE_DAY_LOC'] > pm_date) & (df_master['PRODUCTIONDATE_DAY_LOC'] <= (pm_date + pd.Timedelta(days=7)))]['DOWNTIME'].sum()
        results.append({'DOWNTIME_7D_BEFORE': downtime_before, 'DOWNTIME_7D_AFTER': downtime_after})

    pm_results_df = pd.DataFrame(results)
    summary_data = {
        'Chỉ số': ['Số ca PM phân tích', 'DT trung bình 7 ngày TRƯỚC PM', 'DT trung bình 7 ngày SAU PM', 'Tỷ lệ cải thiện (<0)', 'Tỷ lệ xấu đi (>0)'],
        'Giá trị': [
            len(pm_results_df),
            f"{pm_results_df['DOWNTIME_7D_BEFORE'].mean():.2f}",
            f"{pm_results_df['DOWNTIME_7D_AFTER'].mean():.2f}",
            f"{(pm_results_df['DOWNTIME_7D_AFTER'] < pm_results_df['DOWNTIME_7D_BEFORE']).mean():.2%}",
            f"{(pm_results_df['DOWNTIME_7D_AFTER'] > pm_results_df['DOWNTIME_7D_BEFORE']).mean():.2%}"
        ]
    }
    display(pd.DataFrame(summary_data))
else:
    print("Không tìm thấy sự kiện PM để phân tích.")
```

**Diễn giải & Insight:**
- **Bảo trì là một canh bạc:** Kết quả cho thấy chiến lược bảo trì hiện tại giống như tung đồng xu. Chỉ có **54%** trường hợp cho thấy sự cải thiện sau khi bảo trì. Đáng báo động hơn, có tới **45%** trường hợp tình hình còn tồi tệ hơn.
- **Hàm ý:** Quy trình bảo trì phòng ngừa (PM) hiện tại không đáng tin cậy. Nó cần được xem xét lại một cách toàn diện, từ danh sách công việc (checklist), kỹ năng của kỹ thuật viên, đến quy trình kiểm tra sau bảo trì.
