````python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('../data/processed/master_data.csv', low_memory=False)
df['PRODUCTIONDATE_DAY_LOC'] = pd.to_datetime(df['PRODUCTIONDATE_DAY_LOC'])
df['Total_Production'] = df['GOOD_PRODUCTION_QTY'] + df['REJECT_PRODUCTION_QTY']

# Grouping key
group_cols = ['PRODUCTIONDATE_DAY_LOC', 'SHIFT_NAME', 'LINE_NAME']

# Aggregate base quantities
time_and_qty_cols = [
    'PRODUCTION_AVAILABLE_TIME', 'RUN_TIME', 'GOOD_PRODUCTION_QTY',
    'REJECT_PRODUCTION_QTY', 'Total_Production'
]
agg_base = df.groupby(group_cols)[time_and_qty_cols].sum().reset_index()

# Compute Ideal Time (only where PRODUCTION_ORDER_RATE exists and > 0)
df['Ideal_Time'] = np.where(
    (df['PRODUCTION_ORDER_RATE'].notna()) &
    (df['PRODUCTION_ORDER_RATE'] > 0) &
    (df['Total_Production'] > 0),
    df['Total_Production'] / df['PRODUCTION_ORDER_RATE'],
    0
)
ideal_time = df.groupby(group_cols)['Ideal_Time'].sum().reset_index()

# Merge
agg = pd.merge(agg_base, ideal_time, on=group_cols, how='left')

# Compute OEE components
agg['Availability'] = agg['RUN_TIME'] / np.maximum(agg['PRODUCTION_AVAILABLE_TIME'], 1e-8)
agg['Performance'] = np.where(agg['RUN_TIME'] > 0, agg['Ideal_Time'] / np.maximum(agg['RUN_TIME'], 1e-8), np.nan)
agg['Quality'] = np.where(agg['Total_Production'] > 0, agg['GOOD_PRODUCTION_QTY'] / agg['Total_Production'], np.nan)
agg['OEE'] = agg['Availability'] * agg['Performance'] * agg['Quality']
agg['Defect_Rate'] = np.where(agg['Total_Production'] > 0, agg['REJECT_PRODUCTION_QTY'] / agg['Total_Production'], 0)
agg['Throughput_Yield'] = 1 - agg['Defect_Rate']

# Remove invalid/infinite values
agg = agg.replace([np.inf, -np.inf], np.nan)

# === OUTPUT 1: Overall averages ===
print("=== OVERALL AVERAGES (excluding NaN) ===")
print(f"Average OEE:               {agg['OEE'].mean():.4f} ({agg['OEE'].mean()*100:.2f}%)")
print(f"Average Defect Rate:       {agg['Defect_Rate'].mean():.4f} ({agg['Defect_Rate'].mean()*100:.2f}%)")
print(f"Average Throughput Yield:  {agg['Throughput_Yield'].mean():.4f} ({agg['Throughput_Yield'].mean()*100:.2f}%)")
print()

# === OUTPUT 2: Summary stats (min, max, std) ===
print("=== KPI DISTRIBUTION (OEE, Defect Rate) ===")
summary = agg[['OEE', 'Defect_Rate', 'Throughput_Yield']].describe()
print(summary.round(4))
print()

# === OUTPUT 3: Top 5 worst-performing shifts (by OEE) ===
print("=== TOP 5 WORST SHIFTS BY OEE ===")
worst_oee = agg.nsmallest(5, 'OEE')[['PRODUCTIONDATE_DAY_LOC', 'SHIFT_NAME', 'LINE_NAME', 'OEE', 'Defect_Rate']]
print(worst_oee.to_string(index=False))
print()

# === OUTPUT 4: Top 5 highest defect rates ===
print("=== TOP 5 HIGHEST DEFECT RATES ===")
worst_defect = agg.nlargest(5, 'Defect_Rate')[['PRODUCTIONDATE_DAY_LOC', 'SHIFT_NAME', 'LINE_NAME', 'OEE', 'Defect_Rate']]
print(worst_defect.to_string(index=False))
````

# 🔧 Công thức DAX cho Power BI

Dưới đây là các **measure (biện pháp)** bạn có thể tạo trong **Power BI** để tái tạo các KPI này trực tiếp trong báo cáo, **không cần xử lý trước**.

> 💡 **Giả định:** Bạn đã load file `master_data.csv` vào Power BI với tên bảng là **Production**.

---

## ✅ 1. Tổng sản lượng

```dax
Total Production =
SUM(Production[GOOD_PRODUCTION_QTY]) + SUM(Production[REJECT_PRODUCTION_QTY])
```

---

## ✅ 2. Defect Rate

```dax
Defect Rate =
DIVIDE(
    SUM(Production[REJECT_PRODUCTION_QTY]),
    [Total Production],
    0
)
```

---

## ✅ 3. Throughput Yield

```dax
Throughput Yield =
1 - [Defect Rate]
```

---

## ✅ 4. Tổng thời gian chạy (RUN_TIME)

```dax
Total Run Time =
SUM(Production[RUN_TIME])
```

---

## ✅ 5. Tổng thời gian khả dụng

```dax
Total Available Time =
SUM(Production[PRODUCTION_AVAILABLE_TIME])
```

---

## ✅ 6. Availability

```dax
Availability =
DIVIDE([Total Run Time], [Total Available Time], 0)
```

---

## ✅ 7. Total Ideal Time (phức tạp hơn)

⚠️ **Giải thích:**  
`Ideal Time = Total_Production / PRODUCTION_ORDER_RATE`,  
và `PRODUCTION_ORDER_RATE` chỉ có trên **production events**,  
nên cần dùng **SUMX** hoặc tạo **calculated column**.

**Measure:**

```dax
Total Ideal Time =
SUMX(
    Production,
    IF(
        NOT(ISBLANK(Production[PRODUCTION_ORDER_RATE])) &&
        Production[PRODUCTION_ORDER_RATE] > 0 &&
        (Production[GOOD_PRODUCTION_QTY] + Production[REJECT_PRODUCTION_QTY]) > 0,
        (Production[GOOD_PRODUCTION_QTY] + Production[REJECT_PRODUCTION_QTY]) / Production[PRODUCTION_ORDER_RATE],
        0
    )
)
```

---

## ✅ 8. Performance

```dax
Performance =
DIVIDE([Total Ideal Time], [Total Run Time], 0)
```

---

## ✅ 9. Quality

```dax
Quality =
DIVIDE(SUM(Production[GOOD_PRODUCTION_QTY]), [Total Production], 0)
```

---

## ✅ 10. OEE (cuối cùng!)

```dax
OEE =
[Availability] * [Performance] * [Quality]
```
